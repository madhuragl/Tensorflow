### Creating random tensors

Random tensors are tensors of arbitrary size of random numbers

```python
# Create two random but the same tensor
random_1 = tf.random.Generator.from_seed(7) # set the seed for reproducibility
random_1 = random_1.normal(shape=(3, 2))
random_2 = tf.random.Generator.from_seed(7)
random_2 = random_2.normal(shape=(3, 2))

# Are they equal
random_1, random_2, random_1 == random_2
```

### Shuffle the order of elements in a tensor

```python
# Shuffle a tensor ( valuable when you want to shuffle your data so the inherent order doesnt effect learning)
not_shuffled = tf.constant([[10,7],
                            [3, 4],
                            [2,5]])
#Shuffle our non-shuffled tesnor
tf.random.shuffle(not_shuffled)
```

```python
tf.random.set_seed(42)
tf.random.shuffle(not_shuffled, seed=42)
```

```python
not_shuffled
```

It looks like if we want our shuffled tensors, we would have to do both global and local seed set up

```python
tf.random.set_seed(42)  #global level random seed
tf.random.shuffle(not_shuffled, seed=42)     #operation level random seed
```

**Matrix multiplication**

In Machine Learning, matrix multiplication is one of the most common tensor operations

There are two rules our tensors need to fulfill if we're going to matrix multiply them:
1. The inner dimension must match
2. The resulting matrix has the shape of inner dimensions

```python
#Matrix multiplication in Tensorflow
print(tensor)
tf.matmul(tensor,tensor)
```

```python
tensor * tensor   #element wise
```

```python
# Matrix multiplication with Python operator"@"
tensor @ tensor
```

```python
tensor.shape
```

```python
# Create a tensor (3,2)
X = tf.constant([[1,2],
                 [3,4],
                 [5,6]])
# Create another (3,2) tensor
Y = tf.constant([[7,8],
                 [9,10],
                 [11,12]])
X,Y
```

```python
# Can do the same with Transpose
X, tf.transpose(X), tf.reshape(X, shape=(2,3))
```

**The dot product**

Matrix multiplication is also referred to as dot product

You can perform matrix multiplication using
* tf.matmul()
* tf.tensordot()

```python
X, Y
```

```python
# Perform dot product on x and y.. requires x or y to be transposed
tf.tensordot(tf.transpose(X), Y, axes=1)
```

```python
# Perform matrix multiplication between X and Y (transposed)
tf.matmul(X, tf.transpose(Y))
```

```python
tf.matmul(X, tf.reshape(Y, shape=(2,3)))
```

```python
#check the values of Y, reshape Y and transposed Y
print("Normal Y:")
print(Y, "\n")

print("Y reshaped to (2,3): ")
print(tf.reshape(Y, shape=(2,3)), "\n")

print("Y transposed: ")
print(tf.transpose(Y))
```

```python
tf.matmul(X, tf.transpose(Y))
```

Generally, when performing matrix multiplication on two tensors and one of the axes doesnt line up, you'll transpose rather than reshape, one of the tensors to get satisfy the matrix multiplication
